```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
pd.pandas.set_option('max_columns',None)
```

```python
dataset=pd.read_csv('train.csv')
dataset.head(2)
```

```python
nan_values=[feature for feature in dataset.columns if dataset[feature].isnull().sum()>1 and dataset[feature].dtypes=='O']
for nan in nan_values:
    print(nan,np.round(dataset[nan].isnull().mean(),4),"% missing")
    
```

```python
def nan_replace_with_missing(dataset,nan_values):
    data=dataset.copy()
    data[nan_values]=data[nan_values].fillna('Missing')
    return data
dataset=nan_replace_with_missing(dataset,nan_values)
dataset.isnull().sum()
```

```python
numeric_nan_values=[feature for feature in dataset.columns if dataset[feature].isnull().sum()>1 and dataset[feature].dtypes!='O']
for nan in numeric_nan_values:
    print(nan,np.round(dataset[nan].isnull().mean(),5),"% missing")
```

```python
for feature in numeric_nan_values:
    median_value=dataset[feature].median()
    dataset[feature+'nan']=np.where(dataset[nan].isnull(),1,0)
    dataset.fillna(median_value,inplace=True)
dataset.head(2)
```

```python
for feature in ['YearBuilt','YearRemodAdd','GarageYrBlt']:
    dataset[feature]=dataset['YrSold']-dataset[feature]
dataset.head()
```

```python
dataset[['YearBuilt','YearRemodAdd','GarageYrBlt']].head()
```

```python
dataset.head()
```

```python
num_features=['LotFrontage', 'LotArea', '1stFlrSF', 'GrLivArea', 'SalePrice']

for feature in num_features:
    dataset[feature]=np.log(dataset[feature])
```

```python
dataset.head()
```

```python
categorical_features=[feature for feature in dataset.columns if dataset[feature].dtypes=='O']
for feature in categorical_features:
    temp=dataset.groupby(feature)['SalePrice'].count()/len(dataset)
    temp_df=temp[temp>0.05].index
    dataset[feature]=np.where(dataset[feature].isin(temp_df),dataset[feature],'Rarevar')
```

```python
dataset.head()
```

```python
for feature in categorical_features:
    labels_ordered=dataset.groupby([feature])['SalePrice'].mean().sort_values().index
    labels_ordered={k:i for i,k in enumerate(labels_ordered,0)}
    dataset[feature]=dataset[feature].map(labels_ordered)
```

```python
dataset.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
```

```python
scaling_feature=[feature for feature in dataset.columns if feature not in ['Id','SalePerice'] ]
len(scaling_feature)
```

```python
scaler.fit(dataset[scaling_feature])
```

```python
# transform the train and test set, and add on the Id and SalePrice variables
data = pd.concat([dataset[['Id', 'SalePrice']].reset_index(drop=True),
                    pd.DataFrame(scaler.transform(dataset[scaling_feature]), columns=scaling_feature)],
                    axis=1)
data.head()
```

```python
data.to_csv('new_train.csv',index=False)
```
